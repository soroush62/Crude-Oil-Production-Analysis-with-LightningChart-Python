```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
oil_data = pd.read_excel('../Dataset/DP_LIVE.xlsx')

# Pivot the data to have countries as columns, years as index, and oil production values as values
pivot_data = oil_data.pivot_table(index='TIME', columns='LOCATION', values='Value', aggfunc='sum')

# List of countries to include in the stacked bar chart (modify as necessary)
countries = ['AUS', 'CAN', 'USA']  # Example: Australia, Canada, and USA

# Filter the data to include only the countries of interest
filtered_data = pivot_data[countries]

# Create a stacked bar chart
plt.figure(figsize=(10, 6))

# Plot the data in a stacked manner
filtered_data.plot(kind='bar', stacked=True, color=['skyblue', 'orange', 'green'], ax=plt.gca())

# Add labels and title
plt.title('Crude Oil Production Over Time (Stacked by Country)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Crude Oil Production (KTOE)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Country', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
oil_data = pd.read_excel('../Dataset/DP_LIVE.xlsx')

# Filter the data for a specific year (e.g., 1985)
filtered_data = oil_data[oil_data['TIME'] == 1985]

# Check for missing values and remove rows with NaN in the 'Value' column
year_data = filtered_data.dropna(subset=['Value'])

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(year_data['Value'], labels=year_data['LOCATION'], autopct='%1.1f%%', startangle=90, colors=plt.cm.tab20.colors)

# Add a title
plt.title('Crude Oil Production Breakdown by Country in 1985', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Load your data
oil_data = pd.read_excel('../Dataset/DP_LIVE.xlsx')

# Pivot the data for heatmap visualization (Years as index, Locations as columns)
heatmap_data = oil_data.pivot_table(index='TIME', columns='LOCATION', values='Value', aggfunc='sum')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', linewidths=0.5, annot=False)

# Add labels and title
plt.title('Oil Production Heatmap (Years vs. Locations)', fontsize=14)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Year', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load dataset
file_path = '../Dataset/DP_LIVE.xlsx'
oil_data = pd.read_excel(file_path)

# Preprocess dataset (removing missing/zero values)
oil_data = oil_data[oil_data['Value'] > 0].dropna(subset=['Value'])

# Filter years (1970 - 2017)
oil_data = oil_data[(oil_data['TIME'] >= 1970) & (oil_data['TIME'] <= 2017)]

# Group by 'TIME' to get yearly total production
yearly_data = oil_data.groupby('TIME')['Value'].sum()

# Fit ARIMA model (1,2,1) as identified in analysis
model = ARIMA(yearly_data, order=(1, 2, 1))
model_fit = model.fit()

# Print summary of the model
print(model_fit.summary())

# Forecast future values (for the years 2018-2020)
forecast_values = model_fit.forecast(steps=3)
print("Forecasted values (2018-2020):")
print(forecast_values)

# Plot the forecast results
plt.figure(figsize=(10, 6))
plt.plot(yearly_data.index, yearly_data, label='Historical Data')
plt.plot([2018, 2019, 2020], forecast_values, label='Forecasted Data', color='red')
plt.title('Crude Oil Production Forecast (2018-2020)')
plt.xlabel('Year')
plt.ylabel('Crude Oil Production (KTOE)')
plt.legend()
plt.grid()
plt.show()
```
